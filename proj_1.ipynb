# Web Scraping from Multiple Pages

```python
root = "https://www.football-data.co.uk/mmz4281/"
leagues = ["E0","E1","E2"]
seasons = []
frames = []
for i in range(20,26):
    seasons.append(i)

for league in leagues:
    for season in seasons:
        df = pd.read_csv(root + str(season)+ str(season+1) + "/" + league + ".csv", encoding="latin1").copy()
        df["season"] = season
        df = df[["season"] + [c for c in df.columns if c != "seasons"]]
        frames.append(df)
```
