# adding Coulumn usign .assign() and .insert()

```python
score1 = np.random.randint(0,100,size= 1000)
score2 = np.random.randint(0,100,size= 1000)
```

```python
Series1 = pd.Series(score1,np.arange(0,1000))
Series2 = pd.Series(score2,np.arange(0,1000))

df = df.assign(s1 = Series1,s2 = Series2)
df
```
