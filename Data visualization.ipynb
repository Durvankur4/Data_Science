```python
import pandas as pd
df = pd.read_csv("datasets/population_total.csv")
df.dropna(inplace = True)
df = df.pivot(index = "year",columns = "country")
df.columns = df.columns.get_level_values(-1)
df = df[["United States","India","China","Indonesia","Brazil"]]
```

# Box plot 

```python
# df["India"].plot(kind= "box")
df.plot(kind="box",ylabel = "Population",color ="green")
```

# Histo

```python
df["India"].plot(kind= "hist")
df.plot(kind="hist")
```

# Matplot lib

```python
import matplotlib.pyplot as plt
df.plot(kind="hist")
plt.show()
```
