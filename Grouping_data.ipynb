```python
import pandas as pd
# import numpy as np
```

```python
df_cars = pd.read_csv("datasets/Car_sales.csv")
df_cars.head()
```

```python
df_cars = df_cars[["Manufacturer","Sales_in_thousands",
                   "Vehicle_type","Price_in_thousands",
                   "Engine_size","Horsepower","Fuel_capacity"]]
df_cars.nunique()
```

# agg

```python
df_cars.agg("sum")
```

```python
df_cars.agg({"Sales_in_thousands":["sum","mean","max"],"Price_in_thousands":["sum","max","min"]})
```

# Split apply Combine  

```python
passenger_filter = df_cars["Vehicle_type"]=="Passenger"
car_filter = df_cars["Vehicle_type"] == "Car"
```

```python
avg_passenger= df_cars[passenger_filter]["Sales_in_thousands"].mean()
avg_cars = df_cars[car_filter]["Sales_in_thousands"].mean()
```

```python
pd.DataFrame({"vehicle":["car","pass"],"sales_avg":[avg_passenger,avg_cars]})
```

```python
df_cars.groupby("Vehicle_type").mean(numeric_only=True)
```

```python
df_cars_manu = df_cars.groupby("Manufacturer")
df_cars_manu.groups.keys()
df_cars_manu.get_group("BMW")
```

```python
df_cars.groupby("Manufacturer", as_index = False).mean(numeric_only=True)
```

```python
df_cars.groupby("Vehicle_type").agg(min_sales = ("Sales_in_thousands","min"), max_sales = ("Sales_in_thousands","max"))
```

```python
df_cars.groupby("Manufacturer").sum(numeric_only=True).apply(lambda x : x * 1000)[["Price_in_thousands","Sales_in_thousands"]]
```

```python
num_cols = df_cars.select_dtypes(include= "number").columns
df_cars.groupby("Manufacturer")[num_cols].apply(lambda x : x - x.mean())[["Sales_in_thousands","Price_in_thousands"]]
```

```python
def funct_filter(x) : 
    return x["Sales_in_thousands"].sum()>52
    
df_cars.groupby(["Manufacturer"]).filter(funct_filter)
# df_cars.groupby(["Manufacturer"]).filter(lambda x : x["Sales_in_thousands"].mean()>52 )
```
