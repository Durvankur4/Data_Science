```python
import pandas as pd
df = pd.read_csv("datasets/population_total.csv")
```

```python
df.dropna(inplace = True) 
df = df.pivot(index = "year", columns= "country")
```

```python
print(df.columns)
print(type(df.columns))
df.columns = df.columns.get_level_values(-1)
```

```python
df = df[["India","China","Indonesia","Brazil","United States"]]
```

# Line Plot

```python
df.plot(kind = "line", xlabel= "Year",ylabel ="Population",title= "Population 1950-2020",figsize = (8,5))
```

# Bar plot 

```python
df.tail()
```

```python
df_2020 = df[df.index.isin([2020.0])].T
df_2020.plot(kind="bar",figsize=(4,4))
```
